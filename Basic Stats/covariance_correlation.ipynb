```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def de_mean(x):
    xmean = np.mean(x)
    return [xi - xmean for xi in x]
def covariance(x,y):
    n= len(x)
    y = np.inner(de_mean(x),de_mean(y)) / (n-1)
    return y

pageSpeeds = np.random.normal(3.0,1.0,1000)
purchaseAmount = np.random.normal(50,10.0,1000)
#scatter(pageSpeeds,purchaseAmount)
covariance(pageSpeeds,purchaseAmount)
```

```python
plt.scatter(pageSpeeds,purchaseAmount)
plt.show()
```

```python
#now make purchasae amount real function of page speed
new_purchase_amount = np.random.normal(50.0,10.0,1000)/pageSpeeds
plt.scatter(pageSpeeds,new_purchase_amount)
plt.show()
```

```python
covariance(pageSpeeds,new_purchase_amount)
```

```python
def correlation(x,y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x,y)/ stddevx / stddevy

correlation(pageSpeeds,new_purchase_amount)
    
```

```python
#shortcut
np.corrcoef(pageSpeeds,new_purchase_amount)
np.cov(pageSpeeds,new_purchase_amount)
```

```python
linear_purchase_amount = 100 - pageSpeeds*3
plt.scatter(pageSpeeds,linear_purchase_amount)
plt.show()
```

```python
correlation(pageSpeeds,linear_purchase_amount)
```
