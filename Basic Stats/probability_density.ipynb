```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Uniform Distribution (Constant probbality of every value in range)
values = np.random.uniform(-10.0,10.0,100000)
plt.hist(values,50)
plt.show()
```

```python
from scipy.stats import norm
```

```python
x = np.arange(-3,3,0.001)
plt.plot(x,norm.pdf(x))
```

```python
#mu = mean, sigma = standard deviatoin
mu = 5
sigma = 2
number_of_samples = 10000
values = np.random.normal(mu, sigma, number_of_samples)
plt.hist(values,50)
plt.show()
```

```python
#Exponential PDF or Power Law
from scipy.stats import expon
x = np.arange(0,10,0.01)
plt.plot(x,expon.pdf(x))
```

```python
#Binomail Probability Mass Function
from scipy.stats import binom
n , p = 10,0.5
x = np.arange(0,10,0.01)
plt.plot(x, binom.pmf(x,n,p))
```

```python
#Poisson probability Mass Function
from scipy.stats import poisson
mu = 500
x = np.arange(400,600,0.5)
plt.plot(x,poisson.pmf(x,mu))
```
