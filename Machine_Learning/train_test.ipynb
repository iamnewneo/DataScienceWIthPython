```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
np.random.seed(2)
pageSpeeds = np.random.normal(3.0,1.0,100)
purchaseAmount = np.random.normal(50.0,30,100)/pageSpeeds
```

```python
plt.scatter(pageSpeeds,purchaseAmount)
plt.show()
```

```python
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]
```

```python
plt.scatter(trainX, trainY)
```

```python
plt.scatter(testX,testY)
plt.show()
```

```python
x = np.array(trainX)
y = np.array(trainY)
p8 = np.poly1d(np.polyfit(x,y,8))
```

```python
xp = np.linspace(0,7,100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])
plt.scatter(x,y)
plt.plot(xp, p8(xp), c='r')
```

```python
testx = np.array(testX)
testy = np.array(testY)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])
plt.scatter(testx,testy)
plt.plot(xp, p8(xp), c='r')
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(testy,p8(testx))
print("R2 score for test data: ", r2)
```

```python
r2 = r2_score(np.array(trainY),p8(np.array(trainX)))
print("R2 score for training data: ", r2)
```
